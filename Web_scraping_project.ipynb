```python
import bs4 as bs
import urllib.request
import requests
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stops=stopwords.words('english')
```

```python
scrape_data=urllib.request.urlopen('https://en.wikipedia.org/wiki/Deep_learning')
```

```python
article=scrape_data.read()
```

```python
article
```

```python
parsed_article=bs.BeautifulSoup(article)
```

```python
parsed_article
```

```python
print(parsed_article.p)
```

```python
paragrapgh=parsed_article.find_all('p')
paragrapgh
```

```python
print(parsed_article.title)
```

```python
for para in paragrapgh:
  print(para.text)
```

```python
single_para=" "
for para in paragrapgh:
  single_para=single_para+para.text
```

```python
single_para
```

```python
len(single_para)
```

```python
clean_para=single_para.lower()
clean_para
```

```python
clean_para=re.sub("[^a-z]"," ",clean_para)
clean_para
```

```python
clean_sent=nltk.sent_tokenize(clean_para)
clean_sent
```

```python
# len(clean_sent)
words=[nltk.word_tokenize(all_sent) for all_sent in clean_sent]
# words
len(words)
```

```python
clean_words=[w for w in words if w not in stops]
```

```python
len(clean_words)
```

```python
from gensim.models import Word2Vec
```

```python
window_length=3
embedded_dimention=300
number_of_times_word_has_occured=1
number_of_core=4
epochs=10
```

```python
word_two_vec=Word2Vec(clean_words,size=embedded_dimention,
                      window=window_length,min_count=number_of_times_word_has_occured,
                      iter=epochs,workers=number_of_core,sg=1)
```

```python
vocabulary=word_two_vec.wv.vocab
```

```python
vocabulary
```

```python
v1=word_two_vec.wv['problems']
v1
```

```python
similar_words=word_two_vec.most_similar('problem',topn=5)
similar_words
```

```python
word_two_vec.most_similar(positive=['deep','neural'],negative=['learning'],topn=2)
```

```python
# Saving the model to new_model
word_two_vec.save('model.bin')
```
